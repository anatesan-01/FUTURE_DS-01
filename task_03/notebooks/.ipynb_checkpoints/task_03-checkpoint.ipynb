```python
# ================================
# MARKETING FUNNEL ANALYSIS
# ================================

# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)

# For better visualization
sns.set(style="whitegrid")
```

```python
df = pd.read_csv('D:/Future Intern/task_03/data/bank+marketing/bank/bank-full.csv', sep=';')

df.head()
```

```python
# Dataset shape
print("Dataset Shape:", df.shape)

# Column information
df.info()

# Check missing values
df.isnull().sum()
```

```python
# Convert target column to numeric
df['y'] = df['y'].map({'yes': 1, 'no': 0})

# Verify conversion
df['y'].value_counts()
```

```python
# Total customers
total_customers = len(df)

# Total conversions
total_converted = df['y'].sum()

# Conversion rate
conversion_rate = (total_converted / total_customers) * 100

print("Total Customers:", total_customers)
print("Total Conversions:", total_converted)
print("Overall Conversion Rate: {:.2f}%".format(conversion_rate))
```

```python
funnel_data = pd.DataFrame({
    "Stage": ["Total Customers", "Subscribed Customers"],
    "Count": [total_customers, total_converted],
    "Conversion %": [100, conversion_rate]
})

funnel_data
```

```python
contact_conversion = df.groupby('contact')['y'].mean() * 100
contact_conversion = contact_conversion.sort_values(ascending=False)

contact_conversion
```

```python
contact_conversion.plot(kind='bar')
plt.title("Conversion Rate by Contact Type")
plt.ylabel("Conversion Rate (%)")
plt.xlabel("Contact Type")
plt.xticks(rotation=0)
plt.show()
```

```python
month_conversion = df.groupby('month')['y'].mean() * 100
month_conversion = month_conversion.sort_values(ascending=False)

month_conversion
```

```python
month_conversion.plot(kind='bar')
plt.title("Conversion Rate by Month")
plt.ylabel("Conversion Rate (%)")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.show()
```

```python
job_conversion = df.groupby('job')['y'].mean() * 100
job_conversion = job_conversion.sort_values(ascending=False)

job_conversion
```

```python
job_conversion.plot(kind='bar', figsize=(10,5))
plt.title("Conversion Rate by Job")
plt.ylabel("Conversion Rate (%)")
plt.xticks(rotation=45)
plt.show()
```

```python
campaign_analysis = df.groupby('y')['campaign'].mean()
campaign_analysis
```

```python
sns.boxplot(x='y', y='campaign', data=df)
plt.title("Campaign Calls vs Conversion")
plt.xticks([0,1], ['Not Subscribed','Subscribed'])
plt.show()
```

```python
duration_analysis = df.groupby('y')['duration'].mean()
duration_analysis
```

```python
sns.boxplot(x='y', y='duration', data=df)
plt.title("Call Duration vs Conversion")
plt.xticks([0,1], ['Not Subscribed','Subscribed'])
plt.show()
```

```python
poutcome_conversion = df.groupby('poutcome')['y'].mean() * 100
poutcome_conversion = poutcome_conversion.sort_values(ascending=False)

poutcome_conversion
```

```python
poutcome_conversion.plot(kind='bar')
plt.title("Conversion Rate by Previous Campaign Outcome")
plt.ylabel("Conversion Rate (%)")
plt.xticks(rotation=45)
plt.show()
```

```python
drop_off_rate = 100 - conversion_rate
print("Drop-Off Rate: {:.2f}%".format(drop_off_rate))
```

# Key Insights

1. Overall conversion rate is around 11.70%.
2. Cellular contact performs better than telephone.
3. Customers previously marked as "success" have very high conversion.
4. Longer call duration strongly increases subscription probability.
5. Too many repeated campaign calls reduce effectiveness.
