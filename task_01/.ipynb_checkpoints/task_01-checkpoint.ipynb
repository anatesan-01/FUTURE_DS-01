```python
shop_dt=shop_dt.drop(columns='CustomerID')
```

```python
shop_dt=shop_dt.drop_duplicates()
```

```python
shop_dt
```

```python
shop_dt["InvoiceDate"]=pd.to_datetime(shop_dt['InvoiceDate'])
```

```python
shop_dt
```

```python
shop_dt=shop_dt[(shop_dt["Quantity"]>0)&(shop_dt["UnitPrice"]>0)]
```

```python
shop_dt["Description"]=shop_dt["Description"].fillna("Unknown product")
```

```python
shop_dt
```

```python
shop_dt["revenue"] = shop_dt["Quantity"] * shop_dt["UnitPrice"]
```

### **OverAll revenue**

```python
total_revenue=shop_dt["revenue"].sum()
print(total_revenue)
```

### **Total quatity sold**

```python
total_quantity=shop_dt["Quantity"].sum()
print(total_quantity)
```

### **Number of unique products sold**

```python
shop_dt["StockCode"].nunique()
```

### **Top 10 products by revenue**

```python
top_products=shop_dt.groupby("StockCode")["revenue"].sum().sort_values(ascending=False).head(10)
top_products
```

```python
top_countries=shop_dt.groupby("Country")["revenue"].sum().sort_values(ascending=False).head(10)
top_countries
```

### **Revenue over Time**

```python
shop_dt["InvoiceDate"]=pd.to_datetime(shop_dt["InvoiceDate"])
shop_dt["Months"]=shop_dt["InvoiceDate"].dt.to_period("M")
```

### **monthly revenue trend**

```python
monthly_revenue=shop_dt.groupby("Months")["revenue"].sum()
monthly_revenue
```
